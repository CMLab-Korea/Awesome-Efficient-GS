## 2. ë…¼ë¬¸ ë°ì´í„° êµ¬ì¡° ì •ì˜

```python
# ë…¼ë¬¸ ë°ì´í„° êµ¬ì¡°
paper_tree = {
    "SUCCESS-GS": {
        "3. STATIC (67í¸)": {
            "3.1. Parameter Compression (46í¸)": {
                "3.1.1. Pruning": 14,
                "3.1.2. Attribute Pruning": 3,
                "3.1.3. Quantization": 8,
                "3.1.4. Entropy Coding": 4,
                "3.1.5. Structured Compression": 10
            },
            "3.2. Restructuring Compression (21í¸)": {
                "3.2.1. Anchor-based Hierarchical": 9,
                "3.2.2. Neural Network Integration": 4,
                "3.2.3. Geometric Structure-aware": 4
            }
        },
        "4. DYNAMIC (40í¸)": {
            "4.1. Parameter Compression (12í¸)": {
                "4.1.1. Gaussian Pruning": 6,
                "4.1.2. Attribute Pruning": 1,
                "4.1.3. Quantization": 1,
                "4.1.4. Entropy-based": 4
            },
            "4.2. Restructuring Compression (28í¸)": {
                "4.2.1. Anchor-based": 5,
                "4.2.2. Canonical Deformable": 7,
                "4.2.3. LoD Representation": 2,
                "4.2.4. Mixture-of-Experts": 1
            }
        }
    }
}

# ìƒ‰ìƒ íŒ”ë ˆíŠ¸
colors = {
    'root': '#2E86AB',      # ë£¨íŠ¸ ë…¸ë“œ
    'level1': '#A23B72',    # Static/Dynamic
    'level2': '#F18F01',    # Parameter/Restructuring
    'level3': '#C73E1D',    # ì„¸ë¶€ ì¹´í…Œê³ ë¦¬
    'line': '#666666'       # ì—°ê²°ì„ 
}

print("âœ… ë°ì´í„° êµ¬ì¡° ì •ì˜ ì™„ë£Œ!")
print(f"ì´ ì¹´í…Œê³ ë¦¬ ìˆ˜: {len(paper_tree)}")
```

## 3. íŠ¸ë¦¬ ì‹œê°í™” í•¨ìˆ˜ ì •ì˜

```python
def draw_tree():
    """
    ë…¼ë¬¸ ë¶„ë¥˜ íŠ¸ë¦¬ë¥¼ ì‹œê°í™”í•˜ëŠ” í•¨ìˆ˜
    """
    fig, ax = plt.subplots(figsize=(20, 14))
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 100)
    ax.axis('off')

    # ì œëª©
    ax.text(50, 95, 'SUCCESS-GS: Survey Paper Taxonomy',
            ha='center', va='center', fontsize=24, fontweight='bold',
            bbox=dict(boxstyle='round,pad=1', facecolor=colors['root'],
                     edgecolor='black', linewidth=2, alpha=0.9))

    # ë ˆë²¨ 1: Static/Dynamic
    static_x, static_y = 25, 80
    dynamic_x, dynamic_y = 75, 80

    # Static ë°•ìŠ¤
    ax.add_patch(FancyBboxPatch((static_x-10, static_y-3), 20, 6,
                                boxstyle="round,pad=0.5",
                                facecolor=colors['level1'],
                                edgecolor='black', linewidth=2, alpha=0.9))
    ax.text(static_x, static_y, '3. STATIC\n(67í¸)',
            ha='center', va='center', fontsize=14, fontweight='bold', color='white')

    # Dynamic ë°•ìŠ¤
    ax.add_patch(FancyBboxPatch((dynamic_x-10, dynamic_y-3), 20, 6,
                                boxstyle="round,pad=0.5",
                                facecolor=colors['level1'],
                                edgecolor='black', linewidth=2, alpha=0.9))
    ax.text(dynamic_x, dynamic_y, '4. DYNAMIC\n(40í¸)',
            ha='center', va='center', fontsize=14, fontweight='bold', color='white')

    # ë£¨íŠ¸ì—ì„œ ë ˆë²¨1ë¡œ ì—°ê²°ì„ 
    arrow1 = FancyArrowPatch((50, 92), (static_x, static_y+3),
                            arrowstyle='->', mutation_scale=20,
                            linewidth=2, color=colors['line'])
    arrow2 = FancyArrowPatch((50, 92), (dynamic_x, dynamic_y+3),
                            arrowstyle='->', mutation_scale=20,
                            linewidth=2, color=colors['line'])
    ax.add_patch(arrow1)
    ax.add_patch(arrow2)

    # ë ˆë²¨ 2: Parameter/Restructuring (Static)
    param_comp_x, param_comp_y = 15, 62
    restruct_comp_x, restruct_comp_y = 35, 62

    # Parameter Compression (Static)
    ax.add_patch(FancyBboxPatch((param_comp_x-8, param_comp_y-2.5), 16, 5,
                                boxstyle="round,pad=0.3",
                                facecolor=colors['level2'],
                                edgecolor='black', linewidth=1.5, alpha=0.9))
    ax.text(param_comp_x, param_comp_y, 'Parameter\nCompression\n(46í¸)',
            ha='center', va='center', fontsize=11, fontweight='bold', color='white')

    # Restructuring Compression (Static)
    ax.add_patch(FancyBboxPatch((restruct_comp_x-8, restruct_comp_y-2.5), 16, 5,
                                boxstyle="round,pad=0.3",
                                facecolor=colors['level2'],
                                edgecolor='black', linewidth=1.5, alpha=0.9))
    ax.text(restruct_comp_x, restruct_comp_y, 'Restructuring\nCompression\n(21í¸)',
            ha='center', va='center', fontsize=11, fontweight='bold', color='white')

    # Staticì—ì„œ ë ˆë²¨2ë¡œ ì—°ê²°
    ax.plot([static_x, param_comp_x], [static_y-3, param_comp_y+2.5],
            color=colors['line'], linewidth=1.5, linestyle='-')
    ax.plot([static_x, restruct_comp_x], [static_y-3, restruct_comp_y+2.5],
            color=colors['line'], linewidth=1.5, linestyle='-')

    # ë ˆë²¨ 3: ì„¸ë¶€ ì¹´í…Œê³ ë¦¬ (Static - Parameter)
    static_param_categories = [
        ('Pruning', 14, 5),
        ('Attribute\nPruning', 3, 13),
        ('Quantization', 8, 21),
        ('Entropy\nCoding', 4, 29),
        ('Structured\nCompression', 10, 37)
    ]

    for i, (name, count, x_offset) in enumerate(static_param_categories):
        x = x_offset
        y = 45
        width = 6
        height = 4

        ax.add_patch(FancyBboxPatch((x-width/2, y-height/2), width, height,
                                    boxstyle="round,pad=0.2",
                                    facecolor=colors['level3'],
                                    edgecolor='black', linewidth=1, alpha=0.8))
        ax.text(x, y, f'{name}\n({count}í¸)',
                ha='center', va='center', fontsize=9, color='white', fontweight='bold')

        # ì—°ê²°ì„ 
        ax.plot([param_comp_x, x], [param_comp_y-2.5, y+height/2],
                color=colors['line'], linewidth=1, linestyle='--', alpha=0.7)

    # ë ˆë²¨ 3: ì„¸ë¶€ ì¹´í…Œê³ ë¦¬ (Static - Restructuring)
    static_restruct_categories = [
        ('Anchor-based\nHierarchical', 9, 45),
        ('Neural Network\nIntegration', 4, 53),
        ('Geometric\nStructure-aware', 4, 61)
    ]

    for i, (name, count, x_offset) in enumerate(static_restruct_categories):
        x = x_offset
        y = 45
        width = 7
        height = 4

        ax.add_patch(FancyBboxPatch((x-width/2, y-height/2), width, height,
                                    boxstyle="round,pad=0.2",
                                    facecolor=colors['level3'],
                                    edgecolor='black', linewidth=1, alpha=0.8))
        ax.text(x, y, f'{name}\n({count}í¸)',
                ha='center', va='center', fontsize=9, color='white', fontweight='bold')

        # ì—°ê²°ì„ 
        ax.plot([restruct_comp_x, x], [restruct_comp_y-2.5, y+height/2],
                color=colors['line'], linewidth=1, linestyle='--', alpha=0.7)

    # ë ˆë²¨ 2: Parameter/Restructuring (Dynamic)
    param_comp_d_x, param_comp_d_y = 65, 62
    restruct_comp_d_x, restruct_comp_d_y = 85, 62

    # Parameter Compression (Dynamic)
    ax.add_patch(FancyBboxPatch((param_comp_d_x-8, param_comp_d_y-2.5), 16, 5,
                                boxstyle="round,pad=0.3",
                                facecolor=colors['level2'],
                                edgecolor='black', linewidth=1.5, alpha=0.9))
    ax.text(param_comp_d_x, param_comp_d_y, 'Parameter\nCompression\n(12í¸)',
            ha='center', va='center', fontsize=11, fontweight='bold', color='white')

    # Restructuring Compression (Dynamic)
    ax.add_patch(FancyBboxPatch((restruct_comp_d_x-8, restruct_comp_d_y-2.5), 16, 5,
                                boxstyle="round,pad=0.3",
                                facecolor=colors['level2'],
                                edgecolor='black', linewidth=1.5, alpha=0.9))
    ax.text(restruct_comp_d_x, restruct_comp_d_y, 'Restructuring\nCompression\n(28í¸)',
            ha='center', va='center', fontsize=11, fontweight='bold', color='white')

    # Dynamicì—ì„œ ë ˆë²¨2ë¡œ ì—°ê²°
    ax.plot([dynamic_x, param_comp_d_x], [dynamic_y-3, param_comp_d_y+2.5],
            color=colors['line'], linewidth=1.5, linestyle='-')
    ax.plot([dynamic_x, restruct_comp_d_x], [dynamic_y-3, restruct_comp_d_y+2.5],
            color=colors['line'], linewidth=1.5, linestyle='-')

    # ë ˆë²¨ 3: ì„¸ë¶€ ì¹´í…Œê³ ë¦¬ (Dynamic - Parameter)
    dynamic_param_categories = [
        ('Gaussian\nPruning', 6, 55),
        ('Attribute\nPruning', 1, 62),
        ('Quantization', 1, 69),
        ('Entropy-based', 4, 76)
    ]

    for i, (name, count, x_offset) in enumerate(dynamic_param_categories):
        x = x_offset
        y = 45
        width = 6
        height = 4

        ax.add_patch(FancyBboxPatch((x-width/2, y-height/2), width, height,
                                    boxstyle="round,pad=0.2",
                                    facecolor=colors['level3'],
                                    edgecolor='black', linewidth=1, alpha=0.8))
        ax.text(x, y, f'{name}\n({count}í¸)',
                ha='center', va='center', fontsize=9, color='white', fontweight='bold')

        # ì—°ê²°ì„ 
        ax.plot([param_comp_d_x, x], [param_comp_d_y-2.5, y+height/2],
                color=colors['line'], linewidth=1, linestyle='--', alpha=0.7)

    # ë ˆë²¨ 3: ì„¸ë¶€ ì¹´í…Œê³ ë¦¬ (Dynamic - Restructuring)
    dynamic_restruct_categories = [
        ('Anchor-based', 5, 78),
        ('Canonical\nDeformable', 7, 85),
        ('LoD\nRepresentation', 2, 92),
        ('Mixture-of-\nExperts', 1, 99)
    ]

    for i, (name, count, x_offset) in enumerate(dynamic_restruct_categories):
        x = x_offset
        y = 45
        width = 6
        height = 4

        ax.add_patch(FancyBboxPatch((x-width/2, y-height/2), width, height,
                                    boxstyle="round,pad=0.2",
                                    facecolor=colors['level3'],
                                    edgecolor='black', linewidth=1, alpha=0.8))
        ax.text(x, y, f'{name}\n({count}í¸)',
                ha='center', va='center', fontsize=9, color='white', fontweight='bold')

        # ì—°ê²°ì„ 
        ax.plot([restruct_comp_d_x, x], [restruct_comp_d_y-2.5, y+height/2],
                color=colors['line'], linewidth=1, linestyle='--', alpha=0.7)

    # í†µê³„ ì •ë³´ ì¶”ê°€
    stats_text = (
        "ğŸ“Š Total Papers: 107\n"
        "   â€¢ STATIC: 67 papers\n"
        "   â€¢ DYNAMIC: 40 papers\n\n"
        "ğŸ“… Years: 2024-2026\n"
        "ğŸ† Top Venues: CVPR, ECCV, NeurIPS, ICLR"
    )
    ax.text(50, 15, stats_text,
            ha='center', va='center', fontsize=11,
            bbox=dict(boxstyle='round,pad=1', facecolor='lightgray',
                     edgecolor='black', linewidth=1, alpha=0.7),
            family='monospace')

    # ë²”ë¡€
    legend_elements = [
        mpatches.Patch(facecolor=colors['root'], edgecolor='black', label='Survey Paper'),
        mpatches.Patch(facecolor=colors['level1'], edgecolor='black', label='Static/Dynamic'),
        mpatches.Patch(facecolor=colors['level2'], edgecolor='black', label='Compression Type'),
        mpatches.Patch(facecolor=colors['level3'], edgecolor='black', label='Method Category')
    ]
    ax.legend(handles=legend_elements, loc='upper left', fontsize=10,
             framealpha=0.9, edgecolor='black')

    plt.tight_layout()
    return fig

print("âœ… ì‹œê°í™” í•¨ìˆ˜ ì •ì˜ ì™„ë£Œ!")
```
